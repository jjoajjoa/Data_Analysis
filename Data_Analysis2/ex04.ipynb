```python
import pandas as pd
```

```python
df = pd.read_csv('./data/tmdb_5000_credits.csv')
df.head(1)
```

```python
from ast import literal_eval
df['cast'] = df['cast'].apply(literal_eval)
```

```python
cast = df['cast'].iloc[0]
type(cast)
```

```python
cast
```

```python
# cast에서 10명만 데려오기
def get_list(x):
    if isinstance(x, list):
        names = [i['name'] for i in x]
        if len(names) >= 10 :
            names = names[:10]
        return names
    return []
```

```python
df['cast'] = df['cast'].apply(get_list)
```

```python
cast = df['cast'].iloc[0]
cast
```

```python
crew = df['crew'].iloc[0]
crew
```

```python
df['crew'] = df['crew'].apply(literal_eval)
crew = df['crew'].iloc[0]
```

```python
def get_director(x):
    for i in x:
        if i['job'] == 'Director':
            return i['name']
    return ''
```

```python
df['director'] = df['crew'].apply(get_director)
```

```python
df['director'].head()
```

```python
def clean_data(x):
    if isinstance(x, list):
        return [str.lower(i).replace(' ', '') for i in x]
    else:
        if isinstance(x, str):
            return str.lower(x).replace(' ', '')
        else:
            return ''
```

```python
df['cast'] = df['cast'].apply(clean_data)
df['cast']
```

```python
df['director'] = df['director'].apply(clean_data)
df['director']
```

```python
def create_soup(x):
    return ' '.join(x['cast']) + ' '.join(x['director'])
```

```python
df['soup'] = df.apply(create_soup, axis=1)
df['soup'].head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count = CountVectorizer(stop_words='english')
```

```python
matrix = count.fit_transform(df['soup'])
matrix
```

```python
from sklearn.metrics.pairwise import cosine_similarity
sim = cosine_similarity(matrix, matrix)
sim
```
